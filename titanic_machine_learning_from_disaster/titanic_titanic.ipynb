```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
data.info()
```

```python
data1=data.drop(["Name","Ticket","Cabin","Embarked"],axis=1)
```

```python
data1["Sex"] = data1["Sex"].map({"male": 0, "female": 1})
```

```python
data1.head()
```

```python
corr_matrix = data1.corr()
corr_matrix['Survived'].sort_values(ascending=False)
```

```python
data1.info()
```

```python
data1["Age"].fillna(data1["Age"].mean(),inplace=True)
```

```python
x_train=data1.drop(["Survived"],axis=1)
y_train=data1["Survived"]
```

```python
data1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
t_data=pd.read_csv("test.csv")
t_data
```

```python
t_data.info()
```

```python
t_data.drop(["Name","Ticket","Cabin","Embarked"],axis=1,inplace=True)
```

```python
t_data.info()
```

```python
t_data["Sex"] = t_data["Sex"].map({"male": 0, "female": 1})
t_data["Age"].fillna(t_data["Age"].mean(),inplace=True)
t_data["Fare"].fillna(t_data["Fare"].mean(),inplace=True)
```

```python
t_data.info()
```

```python
model.score(x_train,y_train)
```

```python
y_prd=model.predict(t_data)
```

```python
passengerids=t_data["PassengerId"]
```

```python
Survived=pd.Series(y_prd,name='Survived')
```

```python
submission_df=pd.concat([passengerids,Survived],axis=1)
```

```python
submission_df.to_csv('submit.csv',index=False)
```
